Using the following file https://www.kaggle.com/datasets/gregorut/videogamesales

Create and sort the data by Name, Year, and NA_Sales

````python
#import numpy as np
# np.loadtxt("vgsales-2.csv", dtype={'names':('rank', 'name', 'year', 'NA_Sales')})#, 'formats':('i', 'U', 'i', 'f')})

import numpy as np

# Define the data type for the structured array
dtype = [
    ("Rank", float),  # Changed from int to float
    ("Name", "U100"),
    ("Platform", "U10"),
    ("Year", float),  # Changed from int to float
    ("Genre", "U20"),
    ("Publisher", "U50"),
    ("NA_Sales", float),
    ("EU_Sales", float),
    ("JP_Sales", float),
    ("Other_Sales", float),
    ("Global_Sales", float)
]

# Load data from the CSV file
file_path = "vgsales.csv"  # Update this to the actual path

data = np.genfromtxt(file_path, delimiter=",", dtype=dtype, names=True, encoding="utf-8", missing_values="", filling_values=np.nan, invalid_raise=False)

data["Rank"] = np.nan_to_num(data["Rank"], nan=-1).astype(int)  # Replace NaN with -1
data["Year"] = np.nan_to_num(data["Year"], nan=0).astype(int)   # Replace NaN with 0



# Sorting by Name, Year, and NA_Sales
data_sorted = np.sort(data, order=["Name", "Year", "NA_Sales"])

# Print first 10 rows as a sample
print(data_sorted[:10])
````
